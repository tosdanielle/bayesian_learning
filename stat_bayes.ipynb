## Inference for a proportion: Bayesian Approach (week 1)

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import binom
```

```python
# Probability models and priors
p = np.linspace(0.1, 0.9, num=9)
priors = [0.06, 0.06, 0.06, 0.06, 0.52, 0.06, 0.06, 0.06, 0.06]
```

```python
# Plot prior
plt.bar(p, priors, width = 0.09, align = 'center')
plt.xticks(p)
plt.title('Prior')

plt.show()
```

```python
# Calculate likelihood based on binomial distribuition
def binomial_likelihood(k, n, p): 
    likelihood = []
    for probability in p: 
        l = binom.pmf(k, n, probability)
        likelihood.append(l)
    return likelihood
```

```python
likelihood = binomial_likelihood(4,20, p)
likelihood
```

```python
# Plot likelihood
plt.bar(p,likelihood, width = 0.09, align = 'center')
plt.xticks(p)
plt.title('Likelihood')

plt.show()
```

```python
# Calculate posterior
def calculate_posterior(likelihood, priors): 
    posterior = []
    numerator = []
    denominator = 0
    sum_posterior = 0

    for idx, l in enumerate(likelihood): 
        n = l*priors[idx]
        numerator.append(n)
        denominator += n

    for n in numerator: 
        post = n/denominator
        posterior.append(post)
        sum_posterior += post
    return posterior
```

```python
posterior = calculate_posterior(likelihood, priors)
posterior
```

```python
# Plot posterior
plt.bar(p,posterior, width = 0.09, align = 'center')
plt.xticks(p)
plt.title('Posterior')

plt.show()
```

```python
# Plot prior, likelihood and posterior 
plt.figure(figsize=(18,5))
plt.subplot(1, 3, 1)
#p = np.linspace(0.1, 0.9, num=9)
priors = [0.06, 0.06, 0.06, 0.06, 0.52, 0.06, 0.06, 0.06, 0.06]
plt.bar(p, priors, width = 0.09, align = 'center', color='orange')
plt.xticks(p)
plt.title('Prior')
    
plt.subplot(1, 3, 2)
plt.bar(p,likelihood, width = 0.09, align = 'center', color='purple')
plt.xticks(p)
plt.title('Likelihood')

plt.subplot(1, 3, 3)
plt.bar(p,posterior, width = 0.09, align = 'center')
plt.xticks(p)
plt.title('Posterior')

plt.show()
```

## Conjugacy (week 2)

Bayes' rule for continuous variables

\begin{align}
{\pi}^{*}(p | x) = \frac{P(x|p)\pi (p)}{\int_{0}^{1} P(x|p) \pi(p) dp}\\
\end{align}

**Conjugacy** refers to when the **posterior distribuition** is in the **same family** as your prior belief but with new **parameters values**

## Inference on a binomial proportion (week 2)

RU-486 morning after contraceptive   
   
800 women: 1/2 received RU-486 and the other 1/2 the standard treatment.   
**RU-486 treatment $\qquad$ $\qquad$ $\qquad$ $\qquad$ standard treatment**    
$\quad$0 pregnancies  $\qquad$ $\qquad$ $\qquad$  $\qquad$ $\qquad$ 4 pregnancies   
   
No prior knowledge, therefore we use a uniform distribuition, which means that p assume the same value to all possible values. 

```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1.0, 800)
y = beta.pdf(x, 1, 1)
plt.plot(x, y,"xkcd:orchid")
plt.show()
```

From conjugacy, the posterior distribuition will behave as: 
\begin{align}
beta(\alpha+k, \beta+n-k)
\end{align}

We are interested in the number of occured pregnancies, so n=4. None of which occured on the RU-486 treatment, k=0.

```python
y = beta.pdf(x, 1+0, 1+4-0)
plt.plot(x, y,"xkcd:aquamarine")
plt.show()
```

Your prior belief on the probability of winning in a game of poker is Beta(3,2). You play five games and win four.   What is your new posterior mean for the chance of winning the next game?    
\begin{align}
mean = \frac{\alpha}{\alpha+\beta}
\end{align}
beta(3+4, 2+5-4) = beta(7, 3)   
mean = 0.7
